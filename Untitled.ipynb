```python
# ── Section 8: Text Cleaning & Lemmatization ────────────────
tok = WordPunctTokenizer()
wnl = WordNetLemmatizer()
```

```python
negations_dic = {
    "isn't":"is not", "aren't":"are not", "wasn't":"was not", "weren't":"were not",
    "haven't":"have not", "hasn't":"has not", "hadn't":"had not", "won't":"will not",
    "wouldn't":"would not", "don't":"do not", "doesn't":"does not", "didn't":"did not",
    "can't":"can not", "couldn't":"could not", "shouldn't":"should not",
    "mightn't":"might not", "mustn't":"must not"
}
neg_pattern = re.compile(r'\b(' + '|'.join(negations_dic.keys()) + r')\b')
```

```python
def data_cleaner(t):
    t = str(t)
    t = t.replace(r"Äú",'').replace(r'Äù','').replace(r'\u201c','').replace(r'\u201d','')
    t = t.lower()
    t = neg_pattern.sub(lambda x: negations_dic[x.group()], t)
    t = re.sub("[^a-zA-Z0-9]", " ", t)
    tokens = [w for w in tok.tokenize(t) if len(w) > 3]
    lemmas = [wnl.lemmatize(w) for w in tokens]
    return " ".join(lemmas).strip()
```

```python
for df in (train, test_val, test):
    df['reviews.text']  = df['reviews.text'].apply(data_cleaner)
    df['reviews.title'] = df['reviews.title'].apply(data_cleaner)

print("Text cleaning done.")
train[['reviews.text','reviews.title']].head(5)
```

```python
# ── Section 10: TextBlob Polarity Feature ───────────────────
def get_polarity(x):
    return TextBlob(str(x)).polarity + 1   # shift [-1,1] → [0,2]

for df in (train, test_val, test):
    df['polarity'] = df['reviews.text'].apply(get_polarity)

print("Polarity feature added.")
train[['reviews.text','polarity']].head()
```

```python
# ── Section 11: Custom TF-IDF (for Clustering & LDA) ────────
punc       = ['.', ',', '"', "'", '?', '!', ':', ';', '(', ')', '[', ']', '{', '}', '%']
stop_words = sktext.ENGLISH_STOP_WORDS.union(punc)
stemmer    = SnowballStemmer('english')
re_tok     = RegexpTokenizer(r"[a-zA-Z']+")
```

```python
def tokenize_stem(t):
    return [stemmer.stem(w) for w in re_tok.tokenize(t.lower())]

tvec3 = TfidfVectorizer(stop_words=list(stop_words),
                        tokenizer=tokenize_stem,
                        max_features=1000)

combi_all = pd.concat([train, test_val], axis=0, join='outer').reset_index(drop=True)
reviews   = tvec3.fit_transform(combi_all['reviews.text'])
words     = tvec3.get_feature_names_out()
print("Custom TF-IDF fitted. Vocab size:", len(words))
```

```python
# ── Section 12: Handle Class Imbalance ──────────────────────
print("Class distribution:\n", train.sentiment.value_counts())

counts  = train.sentiment.value_counts().sort_index(ascending=False)
count_2 = counts.iloc[0]
count_1 = counts.iloc[1]
count_0 = counts.iloc[2]

class_2 = train[train.sentiment==2]
class_1 = train[train.sentiment==1]
class_0 = train[train.sentiment==0]
```

```python
# Undersampling
class_2_under = class_2.sample(count_1, random_state=42)
train_under   = pd.concat([class_2_under, class_1, class_0]).reset_index(drop=True)
print("\nUndersampled:\n", train_under.sentiment.value_counts())
```

```python
# Oversampling
class_0_over = class_0.sample(count_2, replace=True, random_state=42)
class_1_over = class_1.sample(count_2, replace=True, random_state=42)
train_over   = pd.concat([class_2, class_0_over, class_1_over]).reset_index(drop=True)
print("\nOversampled:\n", train_over.sentiment.value_counts())
```

```python
# ── Section 13: TF-IDF Feature Matrices ─────────────────────
tvec1 = TfidfVectorizer()
tvec2 = TfidfVectorizer()

train_balanced = train_over.copy()

combi_balanced = pd.concat([train_balanced, test_val], axis=0, join='outer').reset_index(drop=True)
tvec1.fit(combi_balanced['reviews.text'])
tvec2.fit(combi_balanced['reviews.title'])

_drop = ['reviews.text','reviews.title','sentiment','senti_score','polarity']
```

```python
def make_features(train_df, val_df):
    tt1   = pd.DataFrame(tvec1.transform(train_df['reviews.text']).toarray())
    tt2   = pd.DataFrame(tvec1.transform(val_df['reviews.text']).toarray())
    th1   = pd.DataFrame(tvec2.transform(train_df['reviews.title']).toarray())
    th2   = pd.DataFrame(tvec2.transform(val_df['reviews.title']).toarray())
    dc_tr = [c for c in _drop if c in train_df.columns]
    dc_vl = [c for c in _drop if c in val_df.columns]
    X_tr  = pd.concat([train_df.reset_index(drop=True).drop(dc_tr,axis=1), tt1, th1], axis=1)
    X_vl  = pd.concat([val_df.reset_index(drop=True).drop(dc_vl,axis=1),   tt2, th2], axis=1)
    return X_tr.values, X_vl.values

x_train, x_val = make_features(train_balanced, test_val)
y_train = train_balanced['sentiment'].values
y_val   = test_val['sentiment'].values
```

```python
pd.DataFrame(x_train).to_csv('Train.csv', index=False)
pd.DataFrame(x_val).to_csv('Test_Val.csv', index=False)
print("x_train:", x_train.shape, "| x_val:", x_val.shape)
```

```python
# ── Section 14: Reusable Evaluation Function ────────────────
model_results = {}

def evaluate_model(name, y_true, y_pred, plot_roc=True):
    print(f"\n{'='*55}\n {name}\n{'='*55}")
    print(confusion_matrix(y_true, y_pred))
    print(classification_report(y_true, y_pred,
          target_names=['Negative','Neutral','Positive']))
    acc  = accuracy_score(y_true, y_pred) * 100
    f1   = f1_score(y_true, y_pred, average='weighted')
    print(f"Accuracy : {acc:.2f}%  |  F1 (weighted): {f1:.4f}")
    auc_s = None
    if plot_roc:
        lb    = LabelBinarizer()
        yb    = lb.fit_transform(y_true)
        pb    = lb.transform(y_pred)
        auc_s = roc_auc_score(yb, pb, average='weighted')
        print(f"ROC-AUC  : {auc_s:.4f}")
        plt.figure(figsize=(8,5))
        for i, cls in enumerate(['Negative','Neutral','Positive']):
            fpr, tpr, _ = roc_curve(yb[:,i], pb[:,i])
            plt.plot(fpr, tpr, lw=2, label=f'{cls} (AUC={auc(fpr,tpr):.2f})')
        plt.plot([0,1],[0,1],'k--',lw=2)
        plt.xlim([0,1]); plt.ylim([0,1.05])
        plt.xlabel('False Positive Rate'); plt.ylabel('True Positive Rate')
        plt.title(f'ROC Curve — {name}')
        plt.legend(loc='lower right'); plt.tight_layout(); plt.show()
    model_results[name] = {'Accuracy': acc, 'F1': f1, 'AUC': auc_s}
    return acc, f1, auc_s
```

```python
# ── Section 15: Logistic Regression ─────────────────────────
lr = LogisticRegression(C=30, class_weight='balanced',
                        solver='sag', max_iter=1000,
                        random_state=42, n_jobs=-1)
lr.fit(x_train, y_train)
y_pred_lr = lr.predict(x_val)
evaluate_model("Logistic Regression", y_val, y_pred_lr)
```

```python
# ── Section 16: Naive Bayes ──────────────────────────────────
nb = MultinomialNB()
nb.fit(x_train, y_train)
y_pred_nb = nb.predict(x_val)
print("Train score:", nb.score(x_train, y_train))
print("Val   score:", nb.score(x_val,   y_val))
evaluate_model("Naive Bayes", y_val, y_pred_nb)
```

```python
# ── Section 17: Random Forest ───────────────────────────────
rf = RandomForestClassifier(n_estimators=400, random_state=42, n_jobs=-1)
rf.fit(x_train, y_train)
y_pred_rf = rf.predict(x_val)
print("Train score:", rf.score(x_train, y_train))
print("Val   score:", rf.score(x_val,   y_val))
evaluate_model("Random Forest", y_val, y_pred_rf)
```

```python
# ── Section 18: XGBoost ─────────────────────────────────────
xgb = XGBClassifier(n_estimators=300, max_depth=6,
                    use_label_encoder=False, eval_metric='mlogloss',
                    random_state=42, n_jobs=-1)
xgb.fit(x_train, y_train)
y_pred_xgb = xgb.predict(x_val)
evaluate_model("XGBoost", y_val, y_pred_xgb)
```

```python
# ── Section 19: SVM ─────────────────────────────────────────
svc = SVC(kernel='linear', class_weight='balanced',
          C=1.0, random_state=42)
svc.fit(x_train, y_train)
y_pred_svc = svc.predict(x_val)
evaluate_model("SVM (Linear)", y_val, y_pred_svc)
```

```python
# ── Section 20: Voting Classifier ───────────────────────────
voting_clf = VotingClassifier(
    estimators=[('lr', lr), ('nb', nb), ('xgb', xgb)],
    voting='hard'
)
voting_clf.fit(x_train, y_train)
y_pred_vote = voting_clf.predict(x_val)
evaluate_model("Voting Classifier (LR+NB+XGB)", y_val, y_pred_vote)
```

```python
# ── Section 21: Stacking Classifier ─────────────────────────
stacking_clf = StackingClassifier(
    estimators=[
        ('lr',  LogisticRegression(C=30, class_weight='balanced',
                                   solver='saga', max_iter=1000,
                                   random_state=42)),
        ('nb',  MultinomialNB()),
        ('rf',  RandomForestClassifier(n_estimators=100,
                                       random_state=42, n_jobs=-1)),
        ('xgb', XGBClassifier(n_estimators=100, max_depth=4,
                               use_label_encoder=False,
                               eval_metric='mlogloss', random_state=42))
    ],
    final_estimator=LogisticRegression(class_weight='balanced',
                                       max_iter=1000, random_state=42),
    cv=3,
    n_jobs=-1
)
stacking_clf.fit(x_train, y_train)
y_pred_stack = stacking_clf.predict(x_val)
evaluate_model("Stacking Classifier", y_val, y_pred_stack)
```

```python
# ── Section 24: LDA Topic Modeling ──────────────────────────
def print_topics(lda_model, vectorizer, n_top_words):
    vocab = vectorizer.get_feature_names_out()
    for topic_idx, topic in enumerate(lda_model.components_):
        print(f"\nTopic #{topic_idx}:")
        print(" ".join([vocab[i] for i in topic.argsort()[:-n_top_words-1:-1]]))
```

```python
lda = LDA(n_components=10, n_jobs=-1, random_state=42)
lda.fit(reviews)
print("=== Topics found via LDA ===")
print_topics(lda, tvec3, n_top_words=10)
```

```python
def predict_sentiment(review_text, review_title=''):
    cleaned_text  = data_cleaner(review_text)
    cleaned_title = data_cleaner(review_title)

    n_tfidf_text  = tvec1_loaded.get_feature_names_out().shape[0]
    n_tfidf_title = tvec2_loaded.get_feature_names_out().shape[0]
    n_total       = best_loaded.n_features_in_
    n_structured  = n_total - n_tfidf_text - n_tfidf_title

    structured  = np.zeros((1, n_structured))
    text_tfidf  = tvec1_loaded.transform([cleaned_text]).toarray()
    title_tfidf = tvec2_loaded.transform([cleaned_title]).toarray()
    features    = np.hstack([structured, text_tfidf, title_tfidf])

    pred = best_loaded.predict(features)[0]
    return label_map[pred]
    
```

```python
print("\n===== PREDICTIONS ON NEW REVIEWS =====")
for text, title in new_reviews:
    result = predict_sentiment(text, title)
    print(f"Review : {text[:60]}")
    print(f"Result : {result}\n")
```
